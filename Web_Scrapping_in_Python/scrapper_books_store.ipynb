# Web Scrapping in Python
Here is the [Website](https://books.toscrape.com/) to be scrapped

```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd 
```

We are going to get a list of all the links for every single book from all the 50 pages of the website.

```python
# get the base URL of the main page
baseurl = "https://books.toscrape.com/"
'''
- we will send a user-agent on every HTTP request
- because if you make GET request using requests then by default the user-agent is Python which might get blocked.
- So, to override that, we will declare a variable which will store our user-agent
'''
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}
```

We need to inspect the page to find where the links are located.

```python
k = requests.get(baseurl).text
soup = BeautifulSoup(k, 'html.parser')
soup
```

```python
productlist = soup.find_all("h3").text
productlist
```

```python
productlinks = []
for product in productlist:
    link = product.find('a').get('title')
    productlinks.append(baseurl + link)
productlinks
```

```python
productlinks = []
for product in productlist:
    link = product.find("a").get("title")
    productlinks.append(baseurl + link)
productlinks
```
